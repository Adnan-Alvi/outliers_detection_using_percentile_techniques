```python
# Capping ---> Winsarization

new_df_cap = df.copy()

new_df_cap['Height'] = np.where(
    df['Height'] >= upper_limit,
    upper_limit,
    np.where(
        df['Height'] <= lower_limit,
        lower_limit,
        df['Height']
    )
)
```
